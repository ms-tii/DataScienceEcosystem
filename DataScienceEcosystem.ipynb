### Examples: Evaluating Arithmetic Expressions in Python

```python
# 1. This is a simple arithmetic expression that multiplies, then adds integers
(4 * 6) + 5   # Result: 29
```

```python
# 2. This converts 300 minutes to hours by dividing by 40
300 / 40      # Result: 7.5
```

```python
# 3. Calculating compound interest for 1 year
principal = 1000
rate = 0.08      # 7%
times = 4        # quarterly
amount = principal * (1 + rate / times) ** times
amount
```

```python
# 4. Converting temperature from Celsius to Fahrenheit
celsius = 37
fahrenheit = (celsius * 9 / 5) + 32
fahrenheit
```

```python
# 5. Computing the area of a circle
radius = 5
pi = 3.14159
area = pi * radius ** 2
area
```

```python
# 6. Calculating average (mean) of a list of numbers
numbers = [10, 20, 30, 40, 50]
average = sum(numbers) / len(numbers)
average
```

```python
# 7. Simple percentage calculation: discount on a product
price = 1200
discount_percent = 15
discount_amount = price * discount_percent / 100
final_price = price - discount_amount
final_price
```
